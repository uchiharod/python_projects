```python
board = [[' ' for _ in range(3)] for _ in range(3)]

def draw_board():
  print('-' * 9)
  for row in board:
    print('|', *row, '|')
  print('-' * 9)

def get_move(player):
  while True:
    move = input(f'Enter the coordinates for {player}: ')
    try:
      row, col = map(int, move.split())
      if not (0 <= row <= 2 and 0 <= col <= 2):
        print('Coordinates should be from 0 to 2')
      elif board[row][col] != ' ':
        print('This cell is already occupied')
      else:
        return row, col
    except ValueError:
      print('You should enter two integers')

def check_win(player):
  # check rows
  for row in board:
    if row == [player] * 3:
      return True
  # check columns
  for col in range(3):
    if board[0][col] == player and board[1][col] == player and board[2][col] == player:
      return True
  # check diagonals
  if board[0][0] == player and board[1][1] == player and board[2][2] == player:
    return True
  if board[0][2] == player and board[1][1] == player and board[2][0] == player:
    return True
  return False

def main():
  player = 'X'
  while True:
    draw_board()
    row, col = get_move(player)
    board[row][col] = player
    if check_win(player):
      print(f'{player} wins')
      break
    player = 'O' if player == 'X' else 'X'
  draw_board()

main()
```
